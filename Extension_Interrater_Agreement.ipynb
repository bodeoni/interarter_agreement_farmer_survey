```python
import numpy as np
import pandas as pd
```

```python
emma = pd.read_csv("Extension_emma.csv", index_col=0)
tochi =pd.read_csv("Extension_tochi.csv",index_col=0)
```

```python
emma.describe()
```

```python
emma= emma.drop('Unnamed: 43', axis =1)
tochi = tochi.drop('Unnamed: 43', axis =1)
```

```python
emma.info()
```

```python
cols= list(emma.columns.values)
```

```python
cols
```

```python
#Remove columns that are not int64
remove = ['A5', 'A7', 'D5_Other']
for i in remove:
    cols.remove(i)
```

```python
cols
```

```python
#import cohen's kappa function
from sklearn.metrics import cohen_kappa_score as ck
```

```python
print(ck(emma.A9_Training,tochi.A9_Training))
```

```python
#Iteratively compare kappa for all columns in datasets from both scorers
score =[]
for i in cols:
    kappa=ck(emma[i],tochi[i])
    score.append(kappa)
```

```python
score
```

```python
df_scores = pd.DataFrame({'Column': cols, 'Score': score})
```

```python
df_scores
```

```python
df_scores.to_csv('Extension officers Interater Agreement.csv')
```
