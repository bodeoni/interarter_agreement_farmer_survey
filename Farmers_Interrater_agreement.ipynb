```python
import pandas as pd
import numpy as np
from sklearn.metrics import cohen_kappa_score as ck
```

```python
tochi = pd.read_csv("Farmers_tochi.csv", index_col=0)
emma = pd.read_csv("Farmers_emma.csv", index_col=0)
```

```python
tochi.isna().sum()
```

```python
tochi=tochi.dropna()
emma= emma.dropna()
```

```python
cols= list(tochi.columns.values)
```

```python
remove=['A5','A6','A8','B2','B4','B8','B9','C2','C4','C5','D2','D5','D6','E2','E4','E6','E7','E8']
for i in remove:
    cols.remove(i)
```

```python
score =[]
for i in cols:
    x= emma[i].astype(np.int64)
    y= tochi[i].astype(np.int64)
    kappa=ck(x,y)
    score.append(kappa)
```

```python
df_scores = pd.DataFrame({'Column': cols, 'Score': score})
```

```python
df_scores.to_csv('Farmers interater agreement1.csv')
```
